```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as pPlot
from wordcloud import WordCloud, STOPWORDS
import numpy as npy
from PIL import Image
dataset = open("/content/movie list.txt", "r").read()
dataset = dataset.upper()
```

```python
maskArray = npy.array(Image.open("/content/video-300x300.jpg"))
cloud = WordCloud(background_color = "#000012", min_font_size=5, colormap='Blues',
max_font_size=70,font_path='/content/LEMONMILK-Medium.otf',max_words = 2000, width=1400, height=900,
collocations=True, mask = maskArray, stopwords = set(STOPWORDS)).generate(dataset)                
```

```python
# plot the WordCloud image                      
plt.figure(figsize = (9, 14), facecolor = None)
plt.imshow(cloud)
plt.axis("off")
```
